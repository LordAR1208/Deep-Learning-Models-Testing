```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('/content/insurance_data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df[['age']],df.bought_insurance,test_size=0.2,random_state=25)
```

```python
X_train_scaled=X_train.copy()
X_train_scaled['age']=X_train_scaled['age']/100

X_test_scaled=X_test.copy()
X_test_scaled['age']=X_test_scaled['age']/100
```

```python
X_train_scaled
```

```python
model=keras.Sequential([
    keras.layers.Dense(1,input_shape=(1,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros') # Change input_shape to (1,)
])
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train_scaled,y_train,epochs=500)
```

```python
model.evaluate(X_test_scaled,y_test)
```
